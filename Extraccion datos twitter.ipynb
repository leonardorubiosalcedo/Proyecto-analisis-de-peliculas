```python
import pandas as pd
import tweepy
```

```python
# Leer llaves
keys = pd.read_csv("keys_demo.csv", header=None, sep='\t')
keys = dict(zip(keys[0],keys[1]))
```

```python
# Autorizar a travÃ©s API
auth = tweepy.OAuthHandler(keys['api_key'], keys['api_key_secret'])
auth.set_access_token(keys['access_token'], keys['access_token_secret'])
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
datos = pd.read_excel("base_final_movies.xlsx")
datos.head()
```

```python
def creacion_twit(tweets):
    t = []
    for tweet in tweets:
        temp = []
        temp.append(tweet.full_text)
        temp.append(tweet.created_at)
        #temp.append("Bogota")
        #temp.append(tweet.user.location)
        #temp.append(tweet.place)
        #temp.append(tweet.coordinates)
        temp.append(tweet.id)
        temp.append(f"https://twitter.com/i/web/status/{tweet.id}")
        t.append(temp)
    return t
```

```python
def buscar_twit(hastag):
    tweets = tweepy.Cursor(api.search,
                           q=[str(hastag)+' -RT'],
                           lang='en',
                           tweet_mode='extended',
                           include_rts=False).items(30)
    campo=creacion_twit(tweets)
    print(f"El hastag es {hastag}")
    return campo
```

```python
dataframe_final =pd.DataFrame([['1', '2','3','4']] , columns=['tweet','date','id','link'])
def dataframe_gen(campo) :
    data = pd.DataFrame(campo)
    data.columns = ['tweet', 'date',  'id', 'link']
    #print(data)
    return data
```

```python
def unir_data(df_1,df_2):
    df = df_1.append(df_2, ignore_index=True)
    return df
```

```python
campo=[]
validador=714
contador=0
conteo_ciclo=1
for filas in datos['name']:
    contador+=1
    if(contador>validador):
        conteo_ciclo+=1
        variable=buscar_twit(filas)
        contador_len=len(variable)
        if(contador_len==0):
            continue
        dataframe_temp=dataframe_gen(variable)
        dataframe_temp["Pelicula"]=filas
        if(conteo_ciclo==1):
            dataframe_final = dataframe_temp
        else:
            #dataframe_final.append(dataframe_temp, ignore_index = True)
            dataframe_final=unir_data(dataframe_temp,dataframe_final)
        print(contador)
#dataframe_final.head()
```
